```python
import pandas as pd
import mysql.connector
```

```python
# MySQL connection

conn = mysql.connector.connect(
    host='127.0.0.1',
    user='root',
    password='sri123',
    database='food_wastage_management'
)
cursor=conn.cursor()
print(cursor)
```

```python
#load providers

providers=pd.read_csv('E:/Project/Food Waste Management/Datasets/providers_data.csv')
for _,rows in providers.iterrows():
    cursor.execute("""
                   INSERT INTO providers VALUES (%s, %s, %s, %s, %s, %s )
                """,tuple(rows))
print(providers.head())
```

```python
#load receviers

receivers=pd.read_csv('E:/Project/Food Waste Management/Datasets/receivers_data.csv')
for _,rows in receivers.iterrows():
    cursor.execute("""
                   INSERT INTO receivers VALUES (%s, %s, %s, %s, %s )
                """,tuple(rows))
print(receivers.head())
```

```python
# Load Food Listings

food = pd.read_csv('E:/Project/Food Waste Management/Datasets/food_listings_data.csv')
food['Expiry_Date'] = pd.to_datetime(food['Expiry_Date']).dt.date
for _, row in food.iterrows():
    cursor.execute("""
        INSERT INTO food_listings VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s)
    """, tuple(row))
print(food.head())
```

```python
# Load Claims

claims = pd.read_csv('E:/Project/Food Waste Management/Datasets/claims_data.csv')
claims['Timestamp'] = pd.to_datetime(claims['Timestamp'])
for _, row in claims.iterrows():
    cursor.execute("""
        INSERT INTO claims VALUES (%s, %s, %s, %s, %s)
    """, tuple(row))
print(claims.head())
```
